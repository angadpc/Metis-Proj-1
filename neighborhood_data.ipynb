```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

income = pd.read_csv('sub-borougharea-medianhouseholdincome2017.csv')
under18 = pd.read_csv('sub-borougharea-householdswithchildrenunder18yearsold.csv')
```

```python
income['2016 Income'] = income['2016']
income = income[['Sub-Borough Area', '2016 Income']]

under18['2016 - Under 18 %'] = under18['2016']
under18 = under18[['Sub-Borough Area', '2016 - Under 18 %']]

df = pd.merge(income, under18, on='Sub-Borough Area')
df
```

```python
import seaborn as sns
sns.set()
```

```python
plt.scatter(df['2016 Income'], df['2016 - Under 18 %'])
plt.xlabel('Median Income in 2016')
plt.ylabel('Percentage of Households with Children Under 18')
plt.title('NYC Neighborhoods')
```

```python
df.sort_values(by='2016 Income', ascending=False)
```

```python
df.sort_values(by='2016 - Under 18 %', ascending=False)
```

```python
over100 = df[df['2016 Income']>100000]
over100.sort_values(by='2016 - Under 18 %', ascending=False)
```

We selected $100k as a minimum to filter out families that would likely not be able to afford an SAT preparation course. From this, we picked the 3 neighborhoods with the highest percentages of households with children under 18: Park Slope/Carroll Gardens, Greenwich Village/Financial District, and Upper East Side.

From this, we picked 7th Avenue Subway for Park Slope, Fulton Street for Greenwich Village/FiDi, and 86th St (456) for Upper East Side.
